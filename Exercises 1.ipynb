#### 7. On the same strand of DNA there are 2 genes. The first includes nucleotides from position 10 to position 20, the second nucleotides from position 30 to position 40. Let us write this:




```
gene1_start, gene1_end = 10, 20
gene2_start, gene2_end = 30, 40
```

Given a variable pos representing an arbitrary position on the DNA strand, write some comparisons to verify if:


a. pos is in the first gene.


```
def is_in_first_gene(position):
  if pos>=gene1_start and pos<=gene1_end:
    return True
  else:
    return False

position_list = [9, 10, 15, 20, 21]#list of position
for pos in position_list:
  print(f'The position {pos}, is in the first gene?\n{"Yes" if is_in_first_gene(pos) else "No"}')
  print("-"*20)
```

b. pos is in the second gene.

```
def is_in_second_gene(pos):
  if pos>=gene2_start and pos<=gene2_end:
    return True
  else:
      return False

position_list = [0, 10, 20, 30, 31, 40, 41, 50]#list of position
for pos in position_list:
  print(f'The position {pos}, Is in the second gene?\n{"Yes" if is_in_second_gene(pos) else "No"}')
  print("-"*20)
```

c. pos is between the start of the first gene and the end of the second.


```
def between_first_till_second(pos):
  if pos>=gene1_start and pos<=gene2_end:
    return True
  else:
    return False

position_list = [0, 10, 20, 30, 31, 40, 41, 50]#list of position
for pos in position_list:
  print(f'The position {pos}, Is between the start of the first gene and the end of the second?\n{"Yes" if between_first_till_second(pos) else "No"}')
  print("-"*20)
```

d. pos is between the start of the first gene and the end of the second, but not in any of the
genes.

```
def between_first_and_second(pos):
  if pos>=gene1_start and pos<=gene2_end:
    if pos>gene1_end and pos<gene2_start:
      return False
    else:
      return True  
  else:
    return False

position_list = [0, 10, 20, 25,30, 31, 40, 41, 50]#list of position
for pos in position_list:
  print(f'The position {pos}, Is between the start of the first gene and the end of the second, but not in any of the genes?\n{"Yes" if between_first_and_second(pos) else "No"}')
  print("-"*20)
```

e. pos is before the start of the first gene or after the end of the second.

```
def less_or_bigger_first_and_second(pos):
  if pos<gene1_start or pos>gene2_end:
    return True
  else:
    return False

position_list = [0, 10, 20, 25,30, 31, 40, 41, 50]#list of position
for pos in position_list:
  print(f'The position {pos}, Is before the start of the first gene or after the end of the second?\n{"Yes" if less_or_bigger_first_and_second(pos) else "No"}')
  print("-"*20)
```

f. pos is inside one of the genes.

```
def inside_first_or_second(pos):
  if pos>=gene1_start and pos<=gene1_end or pos>=gene2_start and pos<=gene2_end:
    return True
  else:
    return False


position_list = [0, 10, 20, 25,30, 31, 40, 41, 50]#list of position
for pos in position_list:
  print(f'The position {pos}, Is inside one of the genes?\n{"Yes" if inside_first_or_second(pos) else "No"}')
  print("-"*20)
```

g. pos is distant no more than 10 nucleotides from the beginning of the first gene.

```
def distant_10_from_start_first(pos):
  if pos>=gene1_start-10 and pos<=gene1_start+10:
    return True
  else:
    return False

position_list = [0,9, 10, 11, 20, 21]#list of position
for pos in position_list:
  print(f'The position {pos}, Is distant no more than 10 nucleotides from the beginning of the first gene?\n{"Yes" if distant_10_from_start_first(pos) else "No"}')
  print("-"*20)
```
